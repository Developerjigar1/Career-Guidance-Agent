```python
import pandas as pd
import re
```

```python
df = pd.read_csv("../data/row_data/career_dataset.csv")
```

```python
def clean_text(text):
    text = text.lower()
    text = re.sub(r'[^a-zA-Z, ]', '', text)
    return text
```

```python
df["skills"] = df["skills"].apply(clean_text)
df["interest"] = df["interest"].apply(clean_text)

df.head()
```

```python
user_skills = input("Enter your skills (comma separated): ")
user_interest = input("Enter your interest: ")

user_profile = clean_text(user_skills + " " + user_interest)
print("User Profile:", user_profile)
```
